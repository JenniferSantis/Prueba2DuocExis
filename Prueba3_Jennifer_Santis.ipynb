```python
def validar_rut(rut):
    rut = rut.replace(".", "").replace("-", "")
    if not rut[:-1].isdigit() or rut[-1] not in "0123456789Kk" or len(rut) < 2:
        return False
    return True

def agregar_persona(datos_personas):
    rut = input("Ingrese el RUT de la persona (formato: XX.XXX.XXX-X): ")
    if not validar_rut(rut):
        print("RUT inválido")
        return

    nacionalidad = input("Ingrese su nacionalidad: ")
    nombre = input("Ingrese su nombre completo: ")
    fecha_nacimiento = input("Ingrese su fecha de nacimiento (formato: DD/MM/AAAA): ")
    estado_civil = input("Ingrese su estado civil: ")

    datos_personas.append({
        "rut": rut,
        "nacionalidad": nacionalidad,
        "nombre": nombre,
        "fecha_nacimiento": fecha_nacimiento,
        "estado_civil": estado_civil
    })
    print("Datos de persona guardados exitosamente")

def buscar_persona(datos_personas):
    rut = input("Ingrese el RUT de la persona a buscar (formato: XX.XXX.XXX-X): ")
    if not validar_rut(rut):
        print("RUT inválido")
        return

    persona_encontrada = None
    for persona in datos_personas:
        if persona["rut"] == rut:
            persona_encontrada = persona
            break

    if persona_encontrada:
        print("Información de la persona:")
        print("RUT:", persona_encontrada["rut"])
        print("Nacionalidad:", persona_encontrada["nacionalidad"])
        print("Nombre:", persona_encontrada["nombre"])
        print("Fecha de Nacimiento:", persona_encontrada["fecha_nacimiento"])
        print("Estado Civil:", persona_encontrada["estado_civil"])
    else:
        print("No se encontró ninguna persona con ese RUT")

def eliminar_persona(datos_personas):
    rut = input("Ingrese el RUT de la persona a eliminar (formato: XX.XXX.XXX-X): ")
    if not validar_rut(rut):
        print("RUT inválido")
        return

    persona_encontrada = None
    for i, persona in enumerate(datos_personas):
        if persona["rut"] == rut:
            persona_encontrada = persona
            break

    if persona_encontrada:
        confirmacion = input("¿Está seguro/a de eliminar a {}? (S/N): ".format(persona_encontrada["nombre"]))
        if confirmacion.lower() == "s":
            datos_personas.pop(i)
            print("Persona eliminada exitosamente")
        else:
            print("Eliminación cancelada")
    else:
        print("No se encontró ninguna persona con ese RUT")

def emitir_certificado_nacimiento(persona):
    print("Certificado de Nacimiento")
    print("Nombre:", persona["nombre"])
    print("RUT:", persona["rut"])
    print("Fecha de Nacimiento:", persona["fecha_nacimiento"])
    print("Nacionalidad:", persona["nacionalidad"])
    print("Este certificado es válido hasta el 31/12/2023")

def emitir_certificado_estado_civil(persona):
    print("Certificado de Estado Civil")
    print("Nombre:", persona["nombre"])
    print("RUT:", persona["rut"])
    print("Estado Civil:", persona["estado_civil"])
    print("Este certificado es válido hasta el 31/12/2023")

def emitir_certificado_pertenencia(persona):
    print("Certificado de Pertenencia")
    print("Nombre:", persona["nombre"])
    print("RUT:", persona["rut"])
    print("Este certificado es válido hasta el 31/12/2023")

def emitir_certificado(datos_personas):
    rut = input("Ingrese el RUT de la persona para emitir el certificado (formato: XX.XXX.XXX-X): ")
    if not validar_rut(rut):
        print("RUT inválido")
        return

    persona_encontrada = None
    for persona in datos_personas:
        if persona["rut"] == rut:
            persona_encontrada = persona
            break

    if persona_encontrada:
        print("Seleccione el tipo de certificado:")
        print("1. Certificado de Nacimiento")
        print("2. Certificado de Estado Civil")
        print("3. Certificado de Pertenencia")

        tipo_certificado = input("Ingrese el número correspondiente al tipo de certificado: ")
        if tipo_certificado == "1":
            emitir_certificado_nacimiento(persona_encontrada)
        elif tipo_certificado == "2":
            emitir_certificado_estado_civil(persona_encontrada)
        elif tipo_certificado == "3":
            emitir_certificado_pertenencia(persona_encontrada)
        else:
            print("Opción inválida")
    else:
        print("No se encontró ninguna persona con ese RUT")

def mostrar_menu():
    print("---- MENÚ ----")
    print("1. Agregar persona")
    print("2. Buscar persona")
    print("3. Eliminar persona")
    print("4. Emitir certificado")
    print("5. Salir")

def main():
    datos_personas = []

    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            agregar_persona(datos_personas)
        elif opcion == "2":
            buscar_persona(datos_personas)
        elif opcion == "3":
            eliminar_persona(datos_personas)
        elif opcion == "4":
            emitir_certificado(datos_personas)
        elif opcion == "5":
            print("¡Hasta luego!  \n Versión de Sistema 1.0.6.1")
            break
        else:
            print("Opción inválida")

if __name__ == "__main__":
    main()
```
